```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import datetime

# ========== Data Preparation ==========
end_date = datetime.date(2025, 1, 3)
start_date = end_date - datetime.timedelta(days=7*365+3*366+2)

# Fetch and process SPY data
spy = yf.download("SPY", start=start_date, end=end_date, auto_adjust=False)['Adj Close']
spy_monthly = spy.resample('M').last().asfreq('M')
spy_monthly['Return'] = spy_monthly.pct_change().shift(-1)
spy_monthly.dropna(inplace=True)
spy_monthly.index = spy_monthly.index + pd.offsets.MonthEnd(0)
spy_monthly.index = pd.to_datetime(spy_monthly.index).to_period('M').to_timestamp('M')

# Fetch Fama-French factors
ff3 = web.DataReader('F-F_Research_Data_Factors', 'famafrench', start_date, end_date)[0]/100
ff5 = web.DataReader('F-F_Research_Data_5_Factors_2x3', 'famafrench', start_date, end_date)[0]/100

# Convert Fama-French PeriodIndex to DatetimeIndex
ff3.index = ff3.index.to_timestamp() + pd.offsets.MonthEnd(0)
ff5.index = ff5.index.to_timestamp() + pd.offsets.MonthEnd(0)

# Merge with aligned indexes
merged_data = pd.merge(spy_monthly, ff3, left_index=True, right_index=True, how='inner')
merged_data = pd.merge(merged_data, ff5[['Mkt-RF','SMB','HML','RMW','CMA']], 
                       left_index=True, right_index=True, how='inner', suffixes=('_FF3', '_FF5'))
merged_data
```

```python
# Define feature sets
feature_sets = {
    'FF3': ['Mkt-RF_FF3', 'SMB_FF3', 'HML_FF3'],
    'FF5': ['Mkt-RF_FF5', 'SMB_FF5', 'HML_FF5', 'RMW', 'CMA']
}

# ========== Model Configuration ==========
models = {
    'LR': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.001)
}

scaler = StandardScaler()
results = []
window_size = 60  # 5 years in months
test_start = merged_data.index[window_size]

# ========== Rolling Window Prediction ==========
for idx in range(window_size, len(merged_data)-1):
    train_data = merged_data.iloc[idx-window_size:idx]
    test_data = merged_data.iloc[idx:idx+1]
    
    for model_name, model in models.items():
        for fset_name, features in feature_sets.items():
            X_train = train_data[features]
            y_train = train_data['Return']
            
            # Scale features for regularized models
            if model_name in ['Ridge', 'Lasso']:
                X_train_scaled = scaler.fit_transform(X_train)
                model.fit(X_train_scaled, y_train)
                X_test_scaled = scaler.transform(test_data[features])
                pred = model.predict(X_test_scaled)[0]
            else:
                model.fit(X_train, y_train)
                pred = model.predict(test_data[features])[0]
                
            actual = test_data['Return'].values[0]
            results.append({
                'Date': test_data.index[0],
                'Model': model_name,
                'Factors': fset_name,
                'Predicted': pred,
                'Actual': actual
            })

# ========== Performance Evaluation ==========
results_df = pd.DataFrame(results)
metrics = []

for (model, fset), group in results_df.groupby(['Model', 'Factors']):
    y_true = group['Actual']
    y_pred = group['Predicted']
    
    metrics.append({
        'Model': model,
        'Factors': fset,
        'RÂ²': r2_score(y_true, y_pred),
        'MAE': mean_absolute_error(y_true, y_pred),
        'MSE': mean_squared_error(y_true, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred))
    })

metrics_df = pd.DataFrame(metrics)
metrics_df
```

```python
# 1. Time Series of Predicted vs Actual Returns
for (model, fset), group in results_df.groupby(['Model', 'Factors']):
    plt.figure(figsize=(10, 4))
    plt.plot(group['Date'], group['Actual'], label='Actual Return')
    plt.plot(group['Date'], group['Predicted'], label='Predicted Return')
    plt.title(f'{model} with {fset}: Predicted vs Actual Returns')
    plt.xlabel('Date')
    plt.ylabel('Return')
    plt.legend()
    plt.tight_layout()
    plt.show()

# 2. Scatter Plot of Predicted vs Actual Returns
for (model, fset), group in results_df.groupby(['Model', 'Factors']):
    plt.figure(figsize=(6, 6))
    plt.scatter(group['Actual'], group['Predicted'])
    # plot 45Â° identity line
    mn = min(group['Actual'].min(), group['Predicted'].min())
    mx = max(group['Actual'].max(), group['Predicted'].max())
    plt.plot([mn, mx], [mn, mx])
    plt.title(f'{model} with {fset}: Predicted vs Actual')
    plt.xlabel('Actual Return')
    plt.ylabel('Predicted Return')
    plt.tight_layout()
    plt.show()

# Prepare labels for metrics plots
metrics_df['Label'] = metrics_df['Model'] + ' (' + metrics_df['Factors'] + ')'

# 3. Bar Chart of RÂ² by Model/Factors
plt.figure(figsize=(8, 4))
plt.bar(metrics_df['Label'], metrics_df['RÂ²'])
plt.title('RÂ² by Model and Factor Set')
plt.xlabel('Model (Factors)')
plt.ylabel('RÂ²')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Bar Chart of RMSE by Model/Factors
plt.figure(figsize=(8, 4))
plt.bar(metrics_df['Label'], metrics_df['RMSE'])
plt.title('RMSE by Model and Factor Set')
plt.xlabel('Model (Factors)')
plt.ylabel('RMSE')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
