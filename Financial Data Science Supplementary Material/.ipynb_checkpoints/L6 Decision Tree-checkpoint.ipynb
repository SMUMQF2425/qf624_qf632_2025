# CART - Classification

**Run this cell first** to import all the needed class and functions

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Load data set

```python
tennis = pd.read_csv("play tennis.csv")
tennis
```

```python
tennis.info()
```

Convert `Play` column to 0 or 1, so that accuracy rate, precision, recall can be calculated later

```python
tennis['Play'] = (tennis['Play'] == "Yes").astype(int)
```

#### Convert the categorical columns
sklearn implementation of CART handles both categorical & continuous target variables; however, **canNOT support categorical input variables**

Hence we convert categorical columns into dummy variables 0 and 1

NOTE: **only run the below cell once**. We do not want to add duplicate dummy columns to the table again and again.

If the data set has many duplicate dummy columns, load csv again and generate dummy columns again

```python
outlook_dummies = pd.get_dummies(tennis['Outlook'], prefix='Outlook') 
temp_dummies = pd.get_dummies(tennis['Temp'], prefix='Temp') 
humidity_dummies = pd.get_dummies(tennis['Humidity'], prefix='Humidity') 
tennis = pd.concat([tennis, outlook_dummies, temp_dummies, humidity_dummies], axis=1) 
display(tennis)
```

NOTE: `Outlook` has 3 values, and here we make 3 dummy variables. We do not take one value as the baseline while keeping only 2 dummy variables. The goal is interpretability of the decision tree, i.e., we prefer to easily see that `Outlook` is Sunny, rather than seeing that `Outlook_Overcast` and `Outlook_Rainy` are both 0 and hence the `Outlook` is Sunny.

#### Prepare x and y

Note that x is a combination of the newly created dummy variables and the original `Windy` column.

```python
from sklearn.model_selection import train_test_split
x = pd.concat([tennis.loc[:,"Outlook_Overcast":"Humidity_Normal"], tennis.loc[:,"Windy"]], axis=1)
y = tennis.loc[:,"Play"]
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2,random_state = 12345)

display(x_train)
display(y_train)
```

#### Exercise

- Create a Decision Tree model named `dtc` from DecisionTreeClassifier class inside sklearn library's tree module
- Use `dtc` to fit the training data set

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

Run below cell to render the tree in an image

```python
fig = plt.figure(figsize=(8,8))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True)
```

# CART - Regression

#### Load data set

```python
data = pd.read_csv('Advertising.csv', index_col=0) 
display(data)
```

#### Prepare x and y

```python
x = data.loc[:, "TV":"Newspaper"]
y = data.loc[:, "Sales"]
```

#### Split data set into training vs. testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=12345)
```

#### Exercise

Create a Decision Tree model from DecisionTreeRegressor class inside sklearn library's tree module

Check out the documentation at https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

- To obtain a deterministic behaviour during fitting, `random_state` has to be fixed to an integer
- If the improvement of the criterion is identical for several splits, one split has to be selected at random
- Hence the best found split may vary across different runs, even if max_features=n_features.

```python
reg = DecisionTreeRegressor(random_state=0)
reg.fit(x_train, y_train)

fig = plt.figure(figsize=(25,20))
_ = plot_tree(reg, feature_names=x_train.columns, filled=True)
```

Let's try to penalize the tree size and prune the tree. Which parameter shall be set?

Check out the documentation at https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

```python
reg = DecisionTreeRegressor(random_state=0, ccp_alpha=0.3)
reg.fit(x_train, y_train)

fig = plt.figure(figsize=(25,20))
_ = plot_tree(reg, feature_names=x_train.columns, filled=True)
```

#### Assess the final CART regressor generated above

```python
y_pred = reg.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

Use mean squared error as metric

```python
print("MSE:",metrics.mean_squared_error(y_test,y_pred))
```

#### Split data set into training vs testing

Hint:
- x is column `LIMIT_BAL` to `PAY_AMT6`
- y is column `default payment next month`

```python
from sklearn.model_selection import train_test_split
x = credit.loc[:,"LIMIT_BAL":"PAY_AMT6"]
y = credit.loc[:,"default payment next month"]
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2,random_state = 12345)

display(x_train)
display(y_train)
```

#### Generate a classifier model and plot a tree
- Use DecisionTreeClassifier class from sklearn library
- Plot the tree
- Try to penalize the tree size using `ccp_alpha`
- make sure fontsize is 10

```python
dtc = DecisionTreeClassifier(ccp_alpha=0.01)
dtc.fit(x_train, y_train)

fig = plt.figure(figsize=(20,20))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True,
              fontsize=10)
```

Check out documentation https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html 

Experiment various parameters to limit the tree size
- `max_depth`
- `min_samples_split`
- `min_samples_leaf`
- `min_impurity_decrease`

```python
dtc = DecisionTreeClassifier(max_depth = 4)
dtc.fit(x_train, y_train)

fig = plt.figure(figsize=(40,20))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True,
              fontsize=10)
```

```python
dtc = DecisionTreeClassifier(min_samples_split = 3000)
dtc.fit(x_train, y_train)

fig = plt.figure(figsize=(20,20))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True,
              fontsize=10)
```

```python
dtc = DecisionTreeClassifier(min_samples_split = 3000, min_samples_leaf = 100)
dtc.fit(x_train, y_train)

fig = plt.figure(figsize=(20,20))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True,
              fontsize=10)
```

```python
dtc = DecisionTreeClassifier(min_samples_split = 3000, min_samples_leaf = 100, min_impurity_decrease = 0.01)
dtc.fit(x_train, y_train)

fig = plt.figure(figsize=(6,6))
_ = plot_tree(dtc, 
              feature_names = x_train.columns,
              class_names = y_train.unique().astype(str),
              filled=True,
              fontsize=10)
```
