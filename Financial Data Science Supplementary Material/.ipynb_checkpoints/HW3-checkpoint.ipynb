```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, roc_curve, roc_auc_score, confusion_matrix,
                             classification_report)
import matplotlib.pyplot as plt
  
# Load dataset
default_of_credit_card_clients = fetch_ucirepo(id=350)
  
# Metadata 
default_of_credit_card_clients.metadata
```

```python
default_of_credit_card_clients.variables
```

```python
X = default_of_credit_card_clients.data.features
X
```

```python
y = default_of_credit_card_clients.data.targets.values.ravel()
y
```

```python
# Data preprocessing
# Identify categorical and numerical features
numerical_features = ['X1', 'X5'] + [f'X{i}' for i in range(12,24)]
categorical_features = ['X2', 'X3', 'X4'] + [f'X{i}' for i in range(6,12)]

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ])

# Train-test split (stratified due to class imbalance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Model training functions
def train_baseline(X_train, y_train):
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(max_iter=5000, random_state=42))
    ])
    pipeline.fit(X_train, y_train)
    return pipeline

def train_regularized(X_train, y_train, penalty='l1'):
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegressionCV(
            Cs=10, penalty=penalty, solver='saga', max_iter=5000,
            cv=10, random_state=42, scoring='accuracy'
        ))
    ])
    pipeline.fit(X_train, y_train)
    return pipeline

# Train models
print("Training baseline model...")
baseline_model = train_baseline(X_train, y_train)
```

```python
print("Training L1 model...")
l1_model = train_regularized(X_train, y_train, penalty='l1')
```

```python
print("Training L2 model...")
l2_model = train_regularized(X_train, y_train, penalty='l2')
```

```python
# Evaluation functions
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]
    
    # Calculate metrics
    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1': f1_score(y_test, y_pred),
        'AUC': roc_auc_score(y_test, y_proba)
    }
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{model['classifier'].__class__.__name__} (AUC = {metrics['AUC']:.2f})")
    
    return metrics

# Evaluate all models
plt.figure(figsize=(8,6))
print("Evaluating Baseline Model:")
baseline_metrics = evaluate_model(baseline_model, X_test, y_test)

print("\nEvaluating L1 Model:")
l1_metrics = evaluate_model(l1_model, X_test, y_test) # Orange: L1

print("\nEvaluating L2 Model:")
l2_metrics = evaluate_model(l2_model, X_test, y_test) # Green: L2

# Finalize ROC plot
plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.show()
```

```python
# Create comparison table
metrics_df = pd.DataFrame([baseline_metrics, l1_metrics, l2_metrics],
                          index=['Baseline', 'L1 Regularized', 'L2 Regularized'])
print("\nPerformance Comparison:")
metrics_df.round(3)
```
