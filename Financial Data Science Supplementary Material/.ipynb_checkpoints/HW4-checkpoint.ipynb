```python
# Import required libraries
import yfinance as yf
import numpy as np
import pandas as pd
import keras
from pyts.image import GramianAngularField
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models, expected_returns, CovarianceShrinkage

# Download AMZN data
amzn = yf.download('AMZN', start='2015-01-01', end='2025-01-01', auto_adjust=False)
amzn['Return'] = amzn['Adj Close'].pct_change()
amzn = amzn.dropna()
```

Q1

```python
# Create GAF images and labels
def create_gaf_images(df, window_size=20):
    returns = df['Return'].values
    images = []
    labels = []
    dates = []
    
    for i in range(len(returns) - window_size - 1):
        window = returns[i:i+window_size]
        
        # Scale the window to [-1, 1]
        scaler = MinMaxScaler(feature_range=(-1, 1))
        window = scaler.fit_transform(window.reshape(-1, 1)).flatten()
        
        # Create GAF image
        gasf = GramianAngularField(method='summation')
        gaf_image = gasf.fit_transform(window.reshape(1, -1))[0]
        
        # Store date and label
        dates.append(df.index[i+window_size])
        labels.append(1 if returns[i+window_size] > 0 else 0)
        images.append(gaf_image)
    
    return np.array(images), np.array(labels), dates

# Generate features and labels
window_size = 20
X, y, dates = create_gaf_images(amzn, window_size)

# Reshape for CNN input
X = X.reshape(X.shape[0], window_size, window_size, 1)

# Split data into train/test sets
X_train, X_test, y_train, y_test, dates_train, dates_test = train_test_split(
    X, y, dates, test_size=0.2, shuffle=False
)

# Build CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', 
                input_shape=(window_size, window_size, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Compile and train model
model.compile(optimizer='adam', 
             loss='binary_crossentropy',
             metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, 
         validation_split=0.1, verbose=1)

# Generate predictions
test_pred = model.predict(X_test).flatten()

# Convert signals to pandas Series with proper datetime index
strategy_signals = pd.Series(
    (test_pred > 0.5).astype(int),
    index=pd.to_datetime(dates_test)  # Convert to proper datetime index
)

# Calculate strategy returns with lagged signals
test_returns = amzn.loc[dates_test, 'Return']
strategy_returns = test_returns * strategy_signals.shift(1)

# Handle NaN values
strategy_returns = strategy_returns.dropna()

# 1. Check for infinite returns
print("Infinite values in strategy_returns:", np.isinf(strategy_returns).sum())

# 2. Replace infinites (if any) with NaN and drop
strategy_returns = strategy_returns.replace([np.inf, -np.inf], np.nan).dropna()

# 3. Check for zero returns
print("Zero return days:", (strategy_returns == 0).sum())

# 4. Verify returns distribution
print("Returns description:\n", strategy_returns.describe())

# 5. Handle edge case of zero variance
if strategy_returns.var() == 0:
    print("Warning: Strategy returns have zero variance")
    # Add small noise to prevent NaN
    strategy_returns += np.random.normal(0, 1e-6, size=len(strategy_returns))

# Mean-Variance Optimization
portfolio_data = pd.DataFrame(strategy_returns, columns=['AMZN'])

# 1. Calculate expected returns with geometric mean
mu = expected_returns.mean_historical_return(
    portfolio_data,
    returns_data=True,
    compounding=True,
    frequency=252
)

# 2. Calculate covariance matrix with proper returns handling
cov_matrix = risk_models.sample_cov(
    portfolio_data, 
    returns_data=True
) * 252

# 3. Ensure numerical stability
cov_matrix = risk_models.fix_nonpositive_semidefinite(cov_matrix, fix_method='spectral')

# 4. Calculate performance metrics
volatility = np.sqrt(cov_matrix.iloc[0,0])
expected_return = mu.iloc[0]

if abs(volatility) < 1e-8:  # Near-zero threshold
    sharpe = 0.0
else:
    sharpe = expected_return / volatility

print("\nFinal Performance Metrics:")
print(f"Expected Annual Return: {expected_return:.4f}")
print(f"Annual Volatility: {volatility:.4f}")
print(f"Sharpe Ratio: {sharpe:.4f}")
```

Q2

```python
# Define core parameters
window_size = 20
sequence_length = 10  # Number of historical GAF images in each sequence

# Modified function to include date tracking
def create_image_sequence_for_lstm(df, window_size=20, sequence_length=10):
    returns = df['Return'].values
    sequences = []
    labels = []
    dates = []
    
    for i in range(len(returns) - window_size - sequence_length - 1):
        sequence = []
        
        for j in range(sequence_length):
            start_idx = i + j
            end_idx = start_idx + window_size
            window = returns[start_idx:end_idx]
            
            # Scale the window to [-1, 1]
            scaler = MinMaxScaler(feature_range=(-1, 1))
            window = scaler.fit_transform(window.reshape(-1, 1)).flatten()
            
            # Create GAF image
            gasf = GramianAngularField(method='summation')
            gaf_image = gasf.fit_transform(window.reshape(1, -1))[0]
            
            sequence.append(gaf_image.flatten())
        
        # Store corresponding date for the label
        label_date = df.index[i + window_size + sequence_length]
        dates.append(label_date)
        labels.append(1 if returns[i + window_size + sequence_length] > 0 else 0)
        sequences.append(sequence)
    
    return np.array(sequences), np.array(labels), dates

# Generate features, labels, and dates
X, y, dates = create_image_sequence_for_lstm(
    amzn,
    window_size=window_size,
    sequence_length=sequence_length
)

# Split data into train/test sets with temporal preservation
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
dates_train, dates_test = dates[:split_idx], dates[split_idx:]

# Build LSTM model
model = Sequential()
model.add(LSTM(64, 
              return_sequences=True,
              input_shape=(sequence_length, window_size*window_size),
              recurrent_dropout=0.2))
model.add(Dropout(0.3))
model.add(LSTM(32, recurrent_dropout=0.2))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
             loss='binary_crossentropy',
             metrics=['accuracy'])

# Add early stopping
early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# Train model
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=64,
    validation_split=0.15,
    shuffle=False,
    callbacks=[early_stop],
    verbose=1
)

# Generate predictions
test_pred = model.predict(X_test).flatten()

# Create trading signals with proper date alignment
strategy_signals = pd.Series(
    (test_pred > 0.5).astype(int),
    index=pd.to_datetime(dates_test)
)

# Calculate strategy returns with 1-day lag
test_returns = amzn.loc[dates_test, 'Return']
strategy_returns = test_returns * strategy_signals.shift(1)

# Clean returns data
strategy_returns = strategy_returns.dropna().replace([np.inf, -np.inf], np.nan).dropna()

# Handle zero variance case
if strategy_returns.var() < 1e-6:
    strategy_returns += np.random.normal(0, 1e-5, size=len(strategy_returns))

# Portfolio optimization
portfolio_data = pd.DataFrame(strategy_returns, columns=['AMZN'])

# Calculate expected returns (annualized)
mu = expected_returns.mean_historical_return(
    portfolio_data,
    returns_data=True,
    compounding=False,
    frequency=252
)

# Calculate covariance matrix (annualized)
cov_matrix = risk_models.sample_cov(
    portfolio_data,
    returns_data=True
) * 252

# Ensure numerical stability
cov_matrix = risk_models.fix_nonpositive_semidefinite(cov_matrix)

# Calculate performance metrics
volatility2 = np.sqrt(cov_matrix.iloc[0,0])
expected_return = mu.iloc[0]
sharpe_ratio = expected_return / volatility if volatility > 1e-6 else 0.0

print("\nFinal Performance Metrics:")
print(f"Expected Annual Return: {expected_return:.4f}")
print(f"Annual Volatility: {volatility2:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```
