## Exercise 1: Generate synthetic data for linear regression

Create a function that generates a synthetic dataset for linear regression. The function should take the number of samples, the slope, the intercept, and noise as arguments, and return a NumPy array of input values (X) and corresponding output values (Y).

```python
import numpy as np

def generate_synthetic_data(n_samples, slope, intercept, noise_std_dev):
    X = np.random.rand(n_samples) * 10  # Features in [0,10]
    noise = np.random.normal(0, noise_std_dev, n_samples)
    Y = slope * X + intercept + noise
    return X, Y

X, Y = generate_synthetic_data(100, 2, 5, 1)
X
```

```python
Y
```

## Exercise 2: Implement simple linear regression using the normal equation

Implement simple linear regression using the normal equation, and fit a model to the synthetic data generated in Exercise 1.

```python
def normal_equation(X, Y):
    X_b = np.c_[np.ones(len(X)), X]  # Add intercept term
    theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ Y
    return theta_best[1], theta_best[0]  # (slope, intercept)

slope_estimate, intercept_estimate = normal_equation(X, Y)
slope_estimate
```

```python
intercept_estimate
```

## Exercise 3: Use scikit-learn for linear regression

Use the scikit-learn library to fit a linear regression model to the synthetic data generated in Exercise 1, and compare the results with the models obtained in Exercise 2.

```python
from sklearn.linear_model import LinearRegression

def sklearn_linear_regression(X, Y):
    model = LinearRegression()
    model.fit(X.reshape(-1,1), Y)  # Requires 2D feature array
    return model.coef_[0], model.intercept_

slope_estimate, intercept_estimate = sklearn_linear_regression(X, Y)
```

## Exercise 4: Evaluate model performance

Create a function that calculates the Mean Squared Error (MSE) and R-squared score for a given set of true output values (Y) and predicted output values (Y_pred).

```python
def evaluate_model(Y, Y_pred):
    mse = np.mean((Y - Y_pred)**2)
    ss_res = np.sum((Y - Y_pred)**2)
    ss_tot = np.sum((Y - np.mean(Y))**2)
    r2 = 1 - (ss_res / ss_tot)
    return mse, r2

# Example usage with normal equation model
Y_pred = slope_estimate * X + intercept_estimate
mse, r2 = evaluate_model(Y, Y_pred)
```

## Plot for comparison analysis

```python
import matplotlib.pyplot as plt

# Generate predictions for plotting regression lines
X_plot = np.linspace(X.min(), X.max(), 100)
Y_pred_normal = slope_estimate * X_plot + intercept_estimate

# sklearn's model
slope_sklearn, intercept_sklearn = sklearn_linear_regression(X, Y)
Y_pred_sklearn = slope_sklearn * X_plot + intercept_sklearn

# Calculate R² for annotation (using normal equation fit)
Y_pred_full = slope_estimate * X + intercept_estimate
ss_res = np.sum((Y - Y_pred_full) ** 2)
ss_tot = np.sum((Y - np.mean(Y)) ** 2)
r2 = 1 - (ss_res / ss_tot)

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', alpha=0.5, label='Data points')
plt.plot(X_plot, Y_pred_normal, color='red', linewidth=2, label='Normal Equation Fit')
plt.plot(X_plot, Y_pred_sklearn, color='green', linestyle='--', linewidth=2, label='sklearn Fit')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear regression: Data and model comparison')
plt.legend()

# Annotate with estimated parameters and R²
plt.text(0.05, 0.95, f'Normal Eq: slope={slope_estimate:.2f}, intercept={intercept_estimate:.2f}\n'
                     f'sklearn: slope={slope_sklearn:.2f}, intercept={intercept_sklearn:.2f}\n'
                     f'R²={r2:.3f}',
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

plt.show()
```
